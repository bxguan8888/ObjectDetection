```python
import json

idlfile = "/Users/boxuanguan/Documents/python/object_detection_temp/Autonomous-Yolo/DATA/training/label.idl"
car = 0
ped = 0
bike = 0
traffic_light = 0
with open(idlfile) as f:
    for line in f:
        jsonload = json.loads(line)
        assert len(jsonload.keys()) == 1, "Only one image per json file"
        for line in jsonload[jsonload.keys()[0]]:
            cls_id = line[4]
            if cls_id == 1:
                car+=1
            elif cls_id == 2:
                ped +=1
            elif cls_id == 3:
                bike+=1
            elif cls_id == 20:
                traffic_light+=1

print "car:%f, ped:%f, bike:%f, light:%f" % (car*1.0/53901, ped*1.0/53901, bike*1.0/53901, traffic_light*1.0/53901) 
```
